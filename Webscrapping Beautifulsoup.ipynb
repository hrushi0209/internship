```python
year=[]
for i in soup.find_all('span',class_='secondaryInfo'):
    year.append( i.text.replace("(","").replace(")",""))
    
print("Year of Release of the top rated 100 Indian Movies :\n")

res = "\n".join("{}".format(x) for x in zip(year))

print(res)
```

```python
 #Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#a)Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
team = []
for i in soup.find_all('span', class_="u-hide-phablet"):
    team.append(i.text)
team = team[0:10]
print("Top 10 ODI teams in men’s cricket are :\n")
res = "\n".join("{}".format(x) for x in zip(team))
print(res)
```

```python
mat = soup.find('td', class_="rankings-block__banner--matches").text

matches = []
for i in soup.find_all('td', class_="table-body__cell u-center-text"):
    matches.append(i.text)
match=matches[::2]
match.insert(0,mat)
match=match[:10]
print("Top 10 ODI teams in men’s cricket along with the records for matches are :")
res = "\n".join("{}\t{}".format(x, y) for x, y in zip(team,match))
print(res)
```

```python
pt = soup.find('td',class_="rankings-block__banner--points").text

points = []
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    points.append(i.text)
points=matches[1::2]
points.insert(0,pt)
points=points[:10]

print("Top 10 ODI teams in men’s cricket along with the records for matches and points are :")
res = "\n".join("{}\t{}\t{}".format(x, y, z) for x, y, z in zip(team,match,points))

print(res)
```

```python
ratings=[]

for i in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
     ratings.append(i.text.replace("\n","").replace(" ",""))
        
        
for i in soup.find_all('td', class_="table-body__cell u-text-right rating"):
    ratings.append(i.text.replace(" ","").replace("\n",""))
    
ratings=ratings[:10]

print("Top 10 ODI teams in men’s cricket along with the records for matches and points are :")

res = "\n".join("{}\t{}\t{}\t{}".format(w,x, y, z) for w, x, y, z in zip(team,match,points,rating))

print(res)
```

```python
odiMenTeams=pd.DataFrame({"Team":team,"Matches":match,"Points":points,"Ratings":ratings})

print("Top 10 ODI teams in men’s cricket along with the records for matches and points are :")
odiMenTeams
```
